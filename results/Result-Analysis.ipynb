## Comparing Total Power between the Schedulers

```python
import matplotlib.pyplot as plt

# 1. Data for the chart
schedulers = ['Default Scheduler', 'Power-Aware Scheduler']
power_consumption = [11795, 9490]
bar_color = 'royalblue'

# 2. Create the figure and axes
# A slightly wider figure helps with label readability
plt.figure(figsize=(8, 6))

# 3. Create the bar chart
# The 'width' parameter controls the width of the bars.
# A smaller value (e.g., 0.5) creates more space between them.
bars = plt.bar(schedulers, power_consumption, color=bar_color, width=0.5)

# 4. Add titles and labels for clarity
plt.ylabel("Cluster Power Consumption (W)", fontsize=12, fontweight='bold')

# Set the y-axis to start at 0 and add some padding at the top for labels
plt.ylim(0, max(power_consumption) * 1.15)

# 5. Add data labels on top of each bar for precise values
for bar in bars:
    yval = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2.0,  # x-position (center of the bar)
        yval + 200,                         # y-position (slightly above the bar)
        f'{int(yval)} W',                   # The text to display (formatted as integer + " W")
        ha='center',                        # Horizontal alignment
        va='bottom',                        # Vertical alignment
        fontsize=12,
    )

# 6. Add a subtle grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 7. Final layout adjustment and display the plot
plt.tight_layout()
plt.show()
```
